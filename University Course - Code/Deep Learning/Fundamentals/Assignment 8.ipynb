### Consider the following data set of Liquor consumption drinks.csv. Write a program in python 

```python
import pandas as pd
```

### 1. To read the file.

```python
data = pd.read_csv('./data/drinks.csv')
data.head()
```

### 2. To convert country name to capital letters.

```python
data['country'] = data['country'].str.upper()
```

```python
data
```

### 3. Total liquor requirement of each type in each Continent

```python
data.groupby('continent').sum()
```

### 4. Country wise type of liquor which is consumed in maximum quantity

```python
data_country = data.groupby('country').sum()

data_country['max_liquor_type'] = data_country.idxmax(axis=1)

data_country['max_liquor'] = data_country[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']].max(axis=1)

data_country[['max_liquor_type', 'max_liquor']]
```
